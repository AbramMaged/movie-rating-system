```python
from abc import ABC, abstractmethod #


# --------- ABSTRACTION ----------
class Media(ABC):
    def __init__(self, title):
        self.title = title
        self._ratings = []   # ENCAPSULATION

    def add_rating(self, rating):
        if 1 <= rating <= 5:
            self._ratings.append(rating)

    def get_average_rating(self):
        if len(self._ratings) == 0:
            return 0
        return sum(self._ratings) / len(self._ratings)

    @abstractmethod
    def get_info(self):
        pass
```

```python
# --------- INHERITANCE ----------
class Movie(Media):
    def __init__(self, title, director):
        super().__init__(title)
        self.director = director
```

```python
# --------- POLYMORPHISM ----------
class Movie(Media):
    def __init__(self, title, director):
        super().__init__(title)
        self.director = director
    def get_info(self):
        return f"Movie: {self.title}, Director: {self.director}"

class Series(Media):
    def __init__(self, title, seasons):
        super().__init__(title)
        self.seasons = seasons
    def get_info(self):
        return f"Series: {self.title}, Seasons: {self.seasons}"
```

```python
class User:
    def __init__(self, name):
        self.name = name

    def rate_media(self, media, rating):
        media.add_rating(rating)
```

```python
class RatingSystem:
    def __init__(self):
        self.media_list = []

    def add_media(self, media):
        self.media_list.append(media)

    def show_ratings(self):
        for media in self.media_list:
            print(media.get_info())
            print("Average Rating:", media.get_average_rating())
            print()
```

```python
# --------- MAIN PROGRAM ----------

system = RatingSystem()

movie1 = Movie("Inception", "Christopher Nolan")
series1 = Series("Breaking Bad", 5)

user1 = User("Ali")

user1.rate_media(movie1, 5)
user1.rate_media(movie1, 4)
user1.rate_media(series1, 5)

system.add_media(movie1)
system.add_media(series1)

system.show_ratings()
```
