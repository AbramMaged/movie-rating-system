```
from abc import ABC, abstractmethod #stands for Abstract Base Class,
# used to create abstract classes and methods in Python


# --------- ABSTRACTION ----------
class Media(ABC): #super class for 2 (movies,series), also tells python that this is abstract class (ABC)
    def __init__(self, title): #cleean way to object creation, the 2__ helps python recoginize it autmatically
        self.title = title
        self._ratings = []   # ENCAPSULATION (single underscore means private)

    def add_rating(self, rating):#5 star ratting
        if 1 <= rating <= 5: #to insure no rating bigger then 5
            self._ratings.append(rating)

    def get_average_rating(self): #insure no problems happen
        if len(self._ratings) == 0: #took a day off to touch some grass and it was this
            return 0
        return sum(self._ratings) / len(self._ratings)

    @abstractmethod
    def get_info(self):
        pass
    #damn i yap alot
```

```
# --------- INHERITANCE ----------
class Movie(Media): #first child
    def __init__(self, title, director):
        super().__init__(title)
        self.director = director
```

```
# --------- POLYMORPHISM ----------
class Movie(Media):
    def __init__(self, title, director):
        super().__init__(title)
        self.director = director
    def get_info(self):
        return f"Movie: {self.title}, Director: {self.director}"

class Series(Media):#second child
    def __init__(self, title, seasons):
        super().__init__(title)
        self.seasons = seasons
    def get_info(self):
        return f"Series: {self.title}, Seasons: {self.seasons}"
```
